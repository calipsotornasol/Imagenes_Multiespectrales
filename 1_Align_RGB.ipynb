This code uses the Python Open CV library to align red, green and blue band images in the format tif
and combines them to produce an aligned RGB image in the format JPG. In addition, this code
transfers the GPS information from the image in the reference band to the aligned RGB image.

Extensions asociated to each name file refers to a diferente band as is detailed below
    
- '*_1.tif' = Blue Band
- '*_2.tif' = Green Band
- '*_3.tif' = Red Band
- '*_4.tif' = Near Infrarred Band
- '*_5.tif' = Rededge Band

```python
##Preamble
%matplotlib inline  
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import traceback
import sys
from PIL import Image, ImageFilter
import PIL
import matplotlib.pyplot as plt

import datetime
now = datetime.datetime.now()

print(now)
```

```python
##Setting directory
dir_set='/tidchile/collaborative/udd/ids/agro/AGRIFRUT_10-02/MICASENSE/0000SET/001'
num_set=dir_set[-2:]
##Creating lists and listing blue band images 
filenames=glob.glob(dir_set+'/*.tif')
filenamesBLUE=[]
filenamesGREEN=[]
filenamesRED=[]
for i in range(len(filenames)):
    if filenames[i][83]=='1':
        filenamesBLUE.append(filenames[i])
        
filenamesBLUE.sort()        
filenamesBLUE
```

```python
##Listing green band images
filenamesGREEN=[]

for i in range(len(filenames)):
    if filenames[i][83]=='2':
        filenamesGREEN.append(filenames[i])
        
filenamesGREEN.sort()         
filenamesGREEN
```

```python
##Listing red band images
filenamesRED=[]

for i in range(len(filenames)):
    if filenames[i][83]=='3':
        
        filenamesRED.append(filenames[i])
        
filenamesRED.sort()          
filenamesRED
```

```python
##Creating folder to save aligned images
if not os.path.exists(dir_set+'_RGB_ALIGNED/'):
    os.makedirs(dir_set+'_RGB_ALIGNED/')
```

```python
##Preparing to align images
#Define motion model
warp_mode = cv2.MOTION_TRANSLATION  #(Options: MOTION_EUCLIDEAN, MOTION_AFFINE and MOTION_HOMOGRAPHY)
# Set the warp matrix to identity.
if warp_mode == cv2.MOTION_HOMOGRAPHY:
    warp_matrix = np.eye(3, 3, dtype=np.float32)
else:
    warp_matrix = np.eye(2, 3, dtype=np.float32)
```

```python
##Set the stopping criteria for the algorithm.
criteria = (cv2.TERM_CRITERIA_EPS | cv2.TERM_CRITERIA_COUNT, 100000,  1e-5)
```

```python
dir_rgb=dir_set+'_RGB_ALIGNED/'

print(dir_rgb)
```

```python
##Aligning images and saving as .JPG

#Creating log file. List images which were not aligned
log = open(dir_rgb + "error_align.log", "w")

for i in range(len(filenamesRED)):
    try:
        print("Processing: " + filenamesBLUE[i][74:82])
        im_blue        =  cv2.imread(filenamesBLUE[i],0);
        im_green       =  cv2.imread(filenamesGREEN[i],0);
        im_red         =  cv2.imread(filenamesRED[i],0);
              
    
        sz = im_blue.shape
        print (sz)
        height = int(sz[0])
        width = sz[1]
    
        im_aligned = np.zeros((height,width,3), dtype=np.uint8 )
        im_aligned[:,:,2] = im_red[:,:]
    
        #Blue (Aligning Blue)
        (cc, warp_matrix) = cv2.findTransformECC (im_red[:,:], im_blue[:,:],warp_matrix, warp_mode, criteria)
    
        im_aligned[:,:,0] = cv2.warpAffine(im_blue[:,:], warp_matrix, (width, height), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP);
    
        #Green (Aligning Green)
        (cc, warp_matrix) = cv2.findTransformECC (im_red[:,:], im_green[:,:],warp_matrix, warp_mode, criteria)
    
        im_aligned[:,:,1] = cv2.warpAffine(im_green[:,:], warp_matrix, (width, height), flags=cv2.INTER_LINEAR + cv2.WARP_INVERSE_MAP);
        cv2.imwrite(dir_rgb + filenamesBLUE[i][74:82]+'_RGB.JPG',im_aligned)
    except Exception: 
        print("Error processing: " + filenamesBLUE[i][74:82])
        log.write("Error processing: " + filenamesBLUE[i][74:82] + '\n')
        continue
```
